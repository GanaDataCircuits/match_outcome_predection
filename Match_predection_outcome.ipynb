```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['map']=le.fit_transform(df['map'])
df['bomb_planted']=le.fit_transform(df['bomb_planted'])
df['round_winner']=le.fit_transform(df['round_winner'])
```

```python
df.info()
```

```python
#Data splitting in ternms of dependent and independednt
x=df.iloc[:,:-1]
y=df['round_winner']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=34)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler #Standardization
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
# This will learn how to separate classes (CT win vs T win)
lda=LinearDiscriminantAnalysis()
```

```python
# Fit the LDA model on training data
# x_train → features like health, money, time_left, etc.
# y_train → target variable (CT wins or T wins)
lda.fit(x_train,y_train)
```

```python
# This projects features onto directions that best separate CT vs T
lda.transform(x_test)
```

```python
# These are weights assigned to each feature in the LDA decision function
# Higher magnitude = more influence in distinguishing CT vs T wins
lda.coef_ # weights (coefficients) of each feature in the LDA equation
```

```python
# Take absolute value of coefficients
# Reason: sign (+/-) only indicates direction, not importance
# We care about how strong the influence is, not the direction
# Apply exponential to amplify differences between features
# This makes highly important features stand out clearly
```

```python
lda_coef=np.exp(np.abs(lda.coef_))
#np.abs: Taking absolute values because direction doesn’t matter, only strength.
#np.exp: exaggerates the differences to highlight importance.
lda_coef
```

```python
# Convert the coefficient array into a 1D array
# LDA returns coefficients in 2D shape (1, n_features)
# Flattening makes it easier to map with feature names
```

```python
lda_coef=lda_coef.flatten()  #used to convert any dimension array into 1d array
lda_coef
```

```python
feature_names=x.columns
feature_names
```

```python
df_features_imp =pd.DataFrame({'Features':feature_names,'imp_value':lda_coef})
```

```python
df_features_imp
```

```python
# Select the top 20 most important features
# nlargest() sorts features based on importance score
# These features contribute the most to CT vs T classification
```

```python
top_20_fea=df_features_imp.nlargest(20,'imp_value')
#nlargest(n,colname) -->used to get top n values based on colname
top_20_fea

#Picks the top 20 most important features that help LDA distinguish between CT and T wins
```

```python
# Extract indices of the top important features
# These indices correspond to column positions in x_train/x_test arrays
```

```python
imp_col=top_20_fea.index #gets rows indices corresponding to imp features
imp_col #These correspond to column positions in xtrain
```

```python
# Example: selecting specific important feature columns by index
# Here indices like 8, 17, 7 may correspond to:
# ct_health, t_money, time_left
# These features strongly influence round outcome
```

```python
x_train[:,[8,17,7]] #(feature_names[8],[17],[7])
#Picks specific feature columns (by their index) from the training dataset
#— maybe because they're in the top 20 important ones.

#in our dataset, ct_health,t_money and time_left heavily affect whether CT or T
#wins a round
```

```python
x_train=x_train[:,imp_col]
x_train
```

```python
x_test=x_test[:,imp_col]
x_test
```

# Logistic Regression 75

```python
from sklearn.linear_model import LogisticRegression
model_lg=LogisticRegression()
model_lg.fit(x_train,y_train)
```

```python
y_pred=model_lg.predict(x_test)
```

```python
from sklearn.metrics import *
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

# Decison Tree

```python
from sklearn.tree import DecisionTreeClassifier
model_dt=DecisionTreeClassifier()
model_dt.fit(x_train,y_train)
```

```python
y_pred=model_dt.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
model_rf=RandomForestClassifier()
model_rf.fit(x_train,y_train)
```

```python
y_pred=model_rf.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
